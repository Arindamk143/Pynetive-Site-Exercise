```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = pd.read_csv('company_sales_data.csv')
pl = a['total_profit'].tolist()
ml = a['month_number'].tolist()
plt.plot(pl,ml, label = 'month-wish data profit in last year')
plt.xlabel('month Number')
plt.ylabel('profit in dollar')
plt.yticks(ml)
plt.title('Company profit Per Month')
plt.xticks([10000,20000,30000,40000,50000])
plt.show()
```
