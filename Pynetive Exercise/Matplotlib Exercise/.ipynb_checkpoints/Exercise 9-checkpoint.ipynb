```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = pd.read_csv('company_sales_data.csv')
```

```python
a
```

```python
MN =a['month_number'].tolist()
BS =a['bathingsoap'].tolist()
FW = a['facewash'].tolist()
f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(MN,BS, label = 'Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
axarr[0].set_title('Sales data of  a Bathingsoap')
axarr[1].plot(MN,FW,label = 'Face Wash Sales Data', color='r', marker='o', linewidth=3)
axarr[1].set_title('Sales data of  a facewash')
plt.xticks(MN)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()

```
