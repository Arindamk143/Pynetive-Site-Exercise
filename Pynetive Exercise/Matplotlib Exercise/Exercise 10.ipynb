```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = pd.read_csv('company_sales_data.csv')
```

```python
a
```

```python
ML = a['month_number'].tolist()
FC = a['facecream'].tolist()
FW = a['facewash'].tolist()
TP = a['toothpaste'].tolist()
BS = a['bathingsoap'].tolist()
SP = a['shampoo'].tolist()
MZ = a['moisturizer'].tolist()
plt.plot([],[],color='m', label='face Cream', linewidth=5)
plt.plot([],[],color='c', label='Face wash', linewidth=5)
plt.plot([],[],color='r', label='Tooth paste', linewidth=5)
plt.plot([],[],color='k', label='Bathing soap', linewidth=5)
plt.plot([],[],color='g', label='Shampoo', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer', linewidth=5)

plt.stackplot(ML, FC, FW, TP, BS, SP, MZ, colors=['m','c','r','k','g','y'])
plt.legend()
plt.title('All products sales data using the stack plot')
plt.xlabel("Month Numbers")
plt.ylabel("seris unit in numbers")
plt.show()
```
