```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = pd.read_csv('company_sales_data.csv')
```

```python
a
```

```python
import matplotlib.style
ML =a['month_number'].tolist()
TP =a['toothpaste'].tolist()
plt.scatter(ML,TP ,c="blue", label = "Toothpast sales Data")
plt.legend()
plt.title("Toothpast Sales Data")
plt.xlabel("Month Numbers")
plt.ylabel("Number of unit sold")
plt.grid(True, linewidth= 1, linestyle="--")
plt.show()
```
