```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = pd.read_csv('company_sales_data.csv')
```

```python
a
```

```python
PL = a['total_profit'].tolist()
ML = a['month_number'].tolist()
plt.plot(PL,ML, label = 'Profit Data of last Year' , color='r' , marker='o',markerfacecolor='k',linestyle='--', linewidth = 3)
plt.title('Company Sales Data Of Last Year')
plt.xlabel('Month Number')
plt.ylabel('Profit In Dollar')
plt.yticks(ML)
plt.xticks([100000,200000,300000,400000,500000])
plt.legend()
plt.show()
```
