```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = pd.read_csv('company_sales_data.csv')
```

```python
a
```

```python
MN = a['month_number'].tolist()
FC = a['facecream'].tolist()
FW = a['facewash'].tolist()
plt.bar(MN, FC, color = 'b',edgecolor = 'black',
        label='FashCream Sales Data')
plt.legend()
plt.bar(MN, FW, color = 'orange', edgecolor = 'black',
        label='Fashwash Sales Data')
plt.legend()
# plt.bar(MN,FC , color = 'b')
# plt.bar(MN,FW, color ='orange')
plt.grid(True, linestyle='--', linewidth=2)
plt.title("Facewash & FaashCream Sales Data")
plt.xlabel("Month Number")
plt.ylabel("Sales Unit Number")
plt.show()
```
