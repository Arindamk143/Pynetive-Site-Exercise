```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = pd.read_csv('company_sales_data.csv')
```

```python
a
```

```python
MN = a['month_number'].tolist()
BS = a['bathingsoap'].tolist()

plt.bar(MN, BS, label ='Bathingsoap sales data')
plt.legend()
plt.xlabel('Month Number')
plt.ylabel('Sales Unit in number')
plt.title('Bathing soap sales data')
plt.grid(True, linestyle='--', linewidth = 1)
plt.savefig("data.png")
plt.show()
```
