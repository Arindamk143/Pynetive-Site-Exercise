```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = pd.read_csv('company_sales_data.csv')
```

```python
a
```

```python
ML = a['month_number'].tolist()
FC = a['facecream'].tolist()
FW = a['facewash'].tolist()
TP = a['toothpaste'].tolist()
BS = a['bathingsoap'].tolist()
SP = a['shampoo'].tolist()
MZ = a['moisturizer'].tolist()


plt.plot(ML,FC, label = 'Facecreem sales Data' , linewidth = 3)
plt.legend()
plt.plot(ML,FW, label = 'Facewash sales Data' , linewidth = 3)
plt.legend()
plt.plot(ML,TP, label = 'Toothpast sales Data' , linewidth = 3)
plt.legend()
plt.plot(ML,BS, label = 'BathingSoap sales Data' , linewidth = 3)
plt.legend()
plt.plot(ML,SP, label = 'Shampoo sales Data' , linewidth = 3)
plt.legend()
plt.plot(ML,MZ, label = 'Moisturizer sales Data' , linewidth = 3)
plt.legend()

plt.title('Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales Unit in Number')
plt.show()
```
